## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_1 = pd.read_csv('data/hist_data.csv')

plt.hist(df_1['aデータ'], bins=30, color='orange', alpha=0.5, label='aデータ')
plt.hist(df_1['bデータ'], bins=30, color='cyan', alpha=0.4, label='bデータ')

plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度 (℃)')
plt.ylabel('度数')

plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_2 = pd.read_csv('data/scatter_data.csv')

df_red = df_2[df_2["class"] == 0]
df_blue = df_2[df_2["class"] == 1]
df_green = df_2[df_2["class"] == 2]

plt.scatter(df_red['length (cm)'], df_red['width (cm)'], color='red', label='class0')
plt.scatter(df_blue['length (cm)'], df_blue['width (cm)'], color='blue', label='class1')
plt.scatter(df_green['length (cm)'], df_green['width (cm)'], color='green', label='class2')

plt.title('クラス0〜2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.grid(True)
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_3 = pd.read_csv('data/line_data.csv')

fig, ax = plt.subplots()

ax.plot(df_3['期間'], df_3['件数'], marker='.', markersize=10)
plt.setp(ax.get_xticklabels(), rotation=45, fontsize=10)

ax.set_title('期間ごとの件数の推移')
ax.set_xlabel('期間')
ax.set_ylabel('件数')

plt.grid(True)
plt.show()
```

```python
df_3 = pd.read_csv('data/line_data.csv')

plt.bar(df_3['期間'], df_3['費用'])
plt.xticks(rotation=45)

plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用 (円)')

plt.grid(True)
plt.show()
```
